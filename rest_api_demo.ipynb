```python
import os
import json
import requests
from dotenv import load_dotenv

# Загрузка переменных окружения из .env файла
load_dotenv()

print("✅ Модули загружены успешно")
print("📝 Убедитесь, что в файле .env указан корректный OMDB_API_KEY")
```

```python
title = input("Введите фильм для поиска (или Enter для 'The Matrix'): ").strip() or "The Matrix"
if not title:
    raise ValueError("Название фильма не может быть пустым")

api_key = os.getenv("OMDB_API_KEY")
if not api_key:
    raise ValueError("Ключ OMDB_API_KEY не найден в .env")
```

```python
try:
    response = requests.get(
        "https://www.omdbapi.com/",
        params={
            "t": title,
            "apikey": api_key
        },
        timeout=10
    )
    response.raise_for_status()
except requests.exceptions.RequestException as e:
    print(f"❌ Ошибка при запросе к API: {e}")
    print(f"⚠️  Проверьте API ключ в файле .env")
    raise
```

```python
try:
    json_data = response.json()
except json.JSONDecodeError:
    print("❌ Ошибка: API вернул невалидный JSON")
    raise

# Проверка ответа API на ошибку
if json_data.get("Response") == "False":
    print(f"❌ Ошибка API: {json_data.get('Error')}")
    raise ValueError(f"API Error: {json_data.get('Error')}")

print("\n" + "="*50)
print(json.dumps(json_data, ensure_ascii=False, indent=2))
print("="*50 + "\n")
```

```python
movie_title = json_data.get("Title", "Неизвестно")
year = json_data.get("Year", "Неизвестно")
genre = json_data.get("Genre", "Неизвестно")
director = json_data.get("Director", "Неизвестно")
imdb_rating = json_data.get("imdbRating", "N/A")
actors = json_data.get("Actors", "Неизвестно")

print(f"📽️  Фильм: '{movie_title}'")
print(f"📅 Год: {year}")
print(f"🎬 Жанр: {genre}")
print(f"👤 Режиссёр: {director}")
print(f"⭐ Рейтинг IMDb: {imdb_rating}")
print(f"🎭 Актёры: {actors}")
```
